# Color Thresholding

```python
cap = cv2.VideoCapture('/home/nupurjhankar/Desktop/CV_project/15_16.mp4')

ret, frame = cap.read()
print('ret =', ret, 'W =', frame.shape[1], 'H =', frame.shape[0], 'channel =', frame.shape[2])


FPS= 25.0
FrameSize=(frame.shape[1], frame.shape[0])
fourcc = cv2.VideoWriter_fourcc(*'MJPG')

out = cv2.VideoWriter('/home/nupurjhankar/Desktop/CV_project/output_video/15_16.mp4', fourcc, FPS, FrameSize, 1)

while(cap.isOpened()):
    ret, frame = cap.read()

    # check for successfulness of cap.read()
    if not ret: break

    RGB = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    lower_white = np.array([100,100,200])
    upper_white = np.array([255,255,255])
    
    mask = cv2.inRange(RGB, lower_white, upper_white)
    res = cv2.bitwise_and(frame,frame, mask= mask)
    res = (255-res)
    frame = res

    # Save the video
    out.write(frame)

    #cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
         break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
img = cv2.imread("/home/nupurjhankar/Desktop/CV_project/temp2/1/3.jpg", 0)

# sobel derivatives
derivX = cv2.Sobel(img, cv2.CV_32F, 1, 0)
derivY = cv2.Sobel(img, cv2.CV_32F, 0, 1)

# orientation and magnitude
orien = cv2.phase(derivX, derivY, angleInDegrees=True)
mag = cv2.magnitude(derivX, derivY)

# thresholding of the magnitude values, play with the thresh value adjust it too your liking
thresh = 50
_, mask = cv2.threshold(mag, thresh, 255, cv2.THRESH_BINARY)

# I used OpenCV imshow instead of matplotlib, so the colors are in BGR (use yours)
red = np.array([0, 0, 255])
cyan = np.array([255, 255, 0])
green = np.array([0, 255, 0])
yellow = np.array([0, 255, 255])

# for the same reason I use np.uint8
image_map = np.zeros((orien.shape[0], orien.shape[1], 3), dtype=np.uint8)

# setting the colors, maybe there is a better way, my numpy skills are rusty
# it checks that magnitude is above the threshold and that the orientation is in range
image_map[ (mask == 255) & (orien < 90) ] = red
image_map[(mask == 255) & (orien > 90) & (orien < 180)] = cyan
image_map[(mask == 255) & (orien > 180) & (orien < 270)] = green
image_map[(mask == 255) & (orien > 270)] = yellow

# just for showing it with opencv, replace it with matplotlib if you prefer
cv2.imwrite("/home/nupurjhankar/Desktop/CV_project/temp2/1.jpg",image_map)
cv2.imshow("frame", image_map)
cv2.waitKey(0)
   
```

# template matching

```python


img = cv2.imread('/home/nupurjhankar/Desktop/CV_project/temp2/template/In_put/1/1.jpg',0)
img2 = img.copy()
template = cv2.imread('/home/nupurjhankar/Desktop/CV_project/temp2/template/In_put/1/template.jpg',0)
w, h = template.shape[::-1]

# All the 6 methods for comparison in a list
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']

for meth in methods:
    img = img2.copy()
    method = eval(meth)

    # Apply template Matching
    res = cv2.matchTemplate(img,template,method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)

    # If the method is TM_SQDIFF or TM_SQDIFF_NORMED, take minimum
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    bottom_right = (top_left[0] + w, top_left[1] + h)

    cv2.rectangle(img,top_left, bottom_right, 255, 2)
    cv2.imshow("img",img)
    cv2.imshow("res",res)
    
    
    
    plt.imshow(res,cmap = 'gray')
    plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
    plt.savefig("/home/nupurjhankar/Desktop/CV_project/temp2/template/op/"+meth+"D.jpg",)
    plt.imshow(img,cmap = 'gray')
    plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
    plt.savefig("/home/nupurjhankar/Desktop/CV_project/temp2/template/op/"+meth+"M.jpg",)
    plt.suptitle(meth)
    
   
    plt.show()
    
```

# Optical_flow

# Image transformation

```python
def WarpImage(frame):
    ax,bx=10.0,100
    ay,by=20.0,120
    img=np.zeros(frame.shape,dtype=frame.dtype)
    rows,cols=img.shape[:2]
    
    for i in range(rows):
        for j in range(cols):
            offset_x=int(ax*math.sin(2*math.pi*i/bx))
            offset_y=int(ay*math.cos(2*math.pi*j/by))
            if i+offset_y<rows and j+offset_x<cols:
                img[i,j]=frame[(i+offset_y)%rows,(j+offset_x)%cols]
            else:
                img[i,j]=0
    return img

def equalizeHistColor(frame):
    # equalize the histogram of color image
    img = cv2.cvtColor(frame, cv2.COLOR_RGB2HSV)  # convert to HSV
    img[:,:,2] = cv2.equalizeHist(img[:,:,2])     # equalize the histogram of the V channel
    return cv2.cvtColor(img, cv2.COLOR_HSV2RGB)   # convert the HSV image back to RGB format


# start video capture
cap = cv2.VideoCapture('/home/nupurjhankar/Desktop/CV_project/40_46.mp4')
ret, frame = cap.read()
print('ret =', ret, 'W =', frame.shape[1], 'H =', frame.shape[0], 'channel =', frame.shape[2])

i=0
FPS= 25.0
FrameSize=(frame.shape[1], frame.shape[0])
fourcc = cv2.VideoWriter_fourcc(*'MJPG')

out = cv2.VideoWriter('/home/nupurjhankar/Desktop/CV_project/output_video/42_46.mp4', fourcc, FPS, FrameSize, 1)
while(cap.isOpened()):
    # Capture frame-by-frame
    ret, frame = cap.read() 
    #frame=cv2.resize(frame,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_AREA)
    
    # Our operations on the frame come here 
    if ret==1:
        #img = WarpImage(frame)
        img = equalizeHistColor(WarpImage(frame))
    else:
        img = equalizeHistColor(frame)
    frame= np.uint8(frame)
        
    # Display the resulting image
    #cv2.imshow('Warped',img)
    cv2.imwrite('/home/nupurjhankar/Desktop/CV_project/temp/'+str(i)+'.jpg',img)
    i+=1
    out.write(img)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # press q to quit
        break
        
# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

# Eye Detection using Harr features

```python

face_casc = cv2.CascadeClassifier("/home/nupurjhankar/Downloads/python-opencv-detect-master/haarcascade_frontalface_alt.xml")
eye_casc=cv2.CascadeClassifier("/home/nupurjhankar/Downloads/python-opencv-detect-master/haarcascade_eye.xml")

color=(0,255,0)
thickness=3
i=0
    
cap = cv2.VideoCapture(-1)
while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()  # ret = 1 if the video is captured; frame is the image
    
    # Our operations on the frame come here 
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    gray = cv2.equalizeHist(gray)
    faces = face_casc.detectMultiScale(gray, scaleFactor=1.1,minNeighbors=3)
    
    img=frame                     # default if face is not found
    for(x,y,w,h) in faces:
        roi_gray=gray[y:y+h,x:x+w]
        roi_color=frame[y:y+h,x:x+w]
        #img=cv2.rectangle(frame, (x, y), (x + w, y + h), color, thickness) # box for face
        eyes=eye_casc.detectMultiScale(roi_gray)
        for(x_eye,y_eye,w_eye,h_eye) in eyes:
            center=(int(x_eye+0.5*w_eye),int(y_eye+0.5*h_eye))
            radius=int(0.3*(w_eye+h_eye))
            img=cv2.circle(roi_color,center,radius,color,thickness)
            #img=cv2.circle(frame,center,radius,color,thickness)
            cv2.imwrite('/home/nupurjhankar/Desktop/CV_project/temp2/10/'+str(i)+'.jpg',img)
            i+=1


    # Display the resulting image
    cv2.imshow('Face Detection Harr',img)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # press q to quit
        break
        
# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
import re

def tryint(s):
    try:
        return int(s)
    except:
        return s

def alphanum_key(s):
    """ Turn a string into a list of string and number chunks.
        "z23a" -> ["z", 23, "a"]
    """
    return [ tryint(c) for c in re.split('([0-9]+)', s) ]

def sort_nicely(l):
    """ Sort the given list in the way that humans expect.
    """
    l.sort(key=alphanum_key)
```

```python
import cv2
import numpy as np
import os
from os.path import isfile, join
pathIn= '/home/nupurjhankar/Desktop/CV_project/temp2/12/'
pathOut = '/home/nupurjhankar/Desktop/CV_project/temp2/sobel.mp4'
fps = 25
frame_array = []
files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
#for sorting the file names properly
#files.sort(key = lambda x: x[5:-4])
sort_nicely(files)
frame_array = []
files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
#for sorting the file names properly
#files.sort(key = lambda x: x[5:-4])
for i in range(len(files)):
    filename=pathIn + files[i]
    #reading each files
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    
    #inserting the frames into an image array
    frame_array.append(img)
out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
for i in range(len(frame_array)):
    # writing to a image array
    out.write(frame_array[i])
out.release()
```
